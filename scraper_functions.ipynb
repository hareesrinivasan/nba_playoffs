```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.basketball-reference.com/leagues/NBA_2020_games-october.html")
html = response.content
soup = BeautifulSoup(html, "html.parser")
```

```python
months = ["october", "november", "december", "january", "february", "march", "april"]
df_list = []
for month in months:
    url = f"https://www.basketball-reference.com/leagues/NBA_2020_games-{month}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    df_month = pd.read_html(str(soup.find_all("table")[0]))[0]
    df_month.columns = ["Date", "Start_Time", "Visitor", "Visitor_Pts", "Home", "Home_Pts", "Del1", "Del_OT", "Del2", "Del3"]
    df_month.loc[df_month["Del_OT"] == "OT", "df_OT"] = 1
    df_month.loc[df_month["Del_OT"].isna(), "df_OT"] = 0
    df_month = df_month.drop(["Del1", "Del2", "Del3", "Del_OT"], axis=1)
    df_list.append(df_month)
        
```

```python
pd.concat(df_list, axis=0)
```
